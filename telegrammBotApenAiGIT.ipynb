Еще нам понадобиться токен с сайта openai
Процесс получения токена на сайте OpenAI
1. Устанавливаем в браузер Любой рабочий VPN (в моем случае это был ZenMate для браузера FireFox, далее подключаем VPN с европейской страной (в моем случае была Германия)
2. Регестрируемся на сайте openAi вводим почту придумываем пароль и доходим до пункта - введите мобильный номер телефона.
3. Ищем сайт который предоставляем номер мобильного номера для получения смс (бесплатные врятли подойдут, так как они уже 100% регистрировались на сайте openAI проверенно на всех 20 номерах одного сайта) был найден сайт https://onlinesim.io/v2/ru/numbers/

И берем в аренду платный номер для получения смс (внимание даже платный номер не гарантирует регистрацию на сайте - у меня 2-й номер прошел регистрацию). Вводим номер телефона в форму и ждем минуты 2-3 (если номер ранее регистрировался на сайте он сразу это скажет придупреждением) на сайте с арендой номера получем смс и вводим в форму на сайте OpenAI (стоимость аренды 1,64$) Поздравляю у Вас есть регистрация на сайте openAi

```python
bot = Bot(token)
dp = Dispatcher(bot)
```

```python
print(openai.Model.list()) #список моделей openAI для ознакомления выбираем модель Chat используется модель text-davinci-003
```

Будем использовать модель Text-davici-3
Вот что пишут про эту модель:
Новая версия Davinci — text-davinci-003 — получила высокие оценки пользователей в решении логических задач, прописывании мотивации персонажей, стилизации текста для конкретной аудитории и создании креативного контента. По словам разработчиков OpenAI, Davinci лучше понимает смысл текста и успешно решает задачи, связанные с поиском причинно-следственных связей

```python
@dp.message_handler() #мониторинг телеграмм бота
async def send(message : types.Message):#создадим асинхронную функцию
  response = openai.Completion.create( #настройки с сайта openAI с модели text-davinci-003 просто копируем, можно поэксперементироватьс с параметрами
  model="text-davinci-003",
  prompt=message.text,
  temperature=0.5, # температура, чем выше температура, тем более непредсказуемый ответ
  max_tokens=1000,
  top_p=1.0,
  frequency_penalty=0.0,
  presence_penalty=0.6,
  stop=["You:"]
)
  
  await message.answer(response['choices'][0]['text'])#фильтруем и вытаскиваем из нейросети текст который она пишет
```

```python
executor.start_polling(dp, skip_updates=True) 
```
